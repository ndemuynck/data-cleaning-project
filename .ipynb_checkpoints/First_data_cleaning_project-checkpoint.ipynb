### Drop columns

```python
del df['state'] # state only has one value
del df['created_at'] # we don't need the creation date for analysis
del df['the_geom'] # we have long lat column
```

### Check datatypes and transform string values into integers

```python
df.dtypes
```

### Consolidate data

```python
for i in df.columns:
    if df[i].dtype == object:
        df[i] = df[i].str.upper()
df.head()
```

### Transforming text and categorical attributes

```python
encoder = LabelEncoder ()

for i in df.columns:
    if df[i].dtype == object:
        df[i] = encoder.fit_transform(df[i])
df.head()
```

### Set index

```python
df.set_index('tree_id')
df.head ()
```
